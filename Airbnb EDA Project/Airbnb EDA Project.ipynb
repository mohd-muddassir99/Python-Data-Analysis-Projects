### Bivariate Analysis

In this we'll analyse the dependency of one variable on another variable

```python
df.columns
```

#### Let's look how price is dependent on neighbourhood

```python
plt.figure(figsize=(10,4))
sns.barplot(data=df, x='neighbourhood_group', y='price', hue='room_type',saturation=1)
plt.xlabel("Neighbourhood_Group", fontsize=14)
plt.ylabel("Price", fontsize=14)
plt.legend(loc=1)
plt.show()
```

#### Let's understand the relation b/w no.of reviews and price


```python
plt.figure(figsize=(8,5))

sns.scatterplot(data=df, x='number_of_reviews', y='price', hue='neighbourhood_group',markers=5)
plt.title('Locality and Review dependency')
plt.xlabel("No_of_reviews", fontsize=14)
plt.ylabel("Price", fontsize=14)
plt.legend(loc=1)
plt.show()
```

We can see here mostly price lie between 250 to 800 dollars and Brooklyn and Manhattan have the highest no of reviews price
