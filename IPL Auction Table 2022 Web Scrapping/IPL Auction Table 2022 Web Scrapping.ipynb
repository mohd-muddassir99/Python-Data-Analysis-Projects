```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url1 = "https://www.iplt20.com/auction/2022"
r = requests.get(url1)
print(r)
```

```python
soup = BeautifulSoup(r.text, "lxml") # ye soup el var h and apan ne upar ke r ko text me conevrt kiya or "lxml" html ka 2nd name h python me
```

```python
print(soup)
```

```python
table = soup.find("table",class_="ih-td-tab auction-tbl")      
#find func h and soup apna var h jisme pura html code rkha hua h and apan waha web pr gye waha ek corner me button the us pr click krke 
#table pr click q ki apan ko table hi chaiye thi and uske source code me gye or ache se dekha konsa h uska source code waha pr tabkle ki width aayi
#then apan ne uski "Class ko " copy kiya
```

```python
print(table)
```

```python
# find_all :- sare ke sare ko find krta h and find only one ko
header = table.find_all("th") #th mtb titleZ or tags html me ttitles ko th se label krte hen
print(header)
```

```python
titles = []
for i in header:
    title = i.text
    titles.append(title)
print(titles)     #ye apan ne loop chalaya or upar ke header me se only text format ko store karaya

df = pd.DataFrame(columns = titles)
df
```

```python
#abhi apan ne header liye apan rows kenge
rows = table.find_all("tr")  #ye apan ne row find kri jo ki html me "tr " se donete hoti h 
rows
```

```python
for i in rows[1:]:
    data = i.find_all("td")
    #print(data)
    row = [tr.text for tr in data]
    #print(row)   #isko print kroge ke messy data ayega 
    l = len(df)
    #print(l)
    df.loc[l] = row
    df["TEAM"] = df["TEAM"].apply(lambda x:x.strip())
print(df)
#NOW YAAHA PR APAN NE DATA POINTS LIYE AND WITH HELP OF LOOP
```
